```python
import mne 

healthy_files = [f'../data/h{i:02}.edf' for i in range(1, 15)]
schizophrenic_files = [f'../data/s{i:02}.edf' for i in range(1, 15)]
```

```python
def load_eeg_data(file_paths):
    raw_data = []
    for file in file_paths:
        raw = mne.io.read_raw_edf(file, preload=True)
        raw_data.append(raw)
    return raw_data
```

```python
healthy_eeg_data = load_eeg_data(healthy_files)
schizophrenic_eeg_data = load_eeg_data(schizophrenic_files)
print(len(healthy_eeg_data), len(schizophrenic_eeg_data))
```

```python
def preprocess(raw_eeg):
    raw_eeg.filter(0.1,40.,fir_design = 'firwin')
    raw_eeg.set_eeg_reference('average',projection=True)
    return raw_eeg 
```

```python
preprocessed_healthy_eeg = [preprocess(raw) for raw in healthy_eeg_data]
preprocessed_schizophrenic_eeg = [
    preprocess(raw) for raw in schizophrenic_eeg_data]
```

```python
def epoch_eeg(raw_eeg):
    events = mne.make_fixed_length_events(raw_eeg, duration=1.0)
    epochs = mne.Epochs(raw_eeg, events, tmin=0, tmax=2,
                        baseline=None, preload=True)
    return epochs
```

```python
healthy_epochs = [epoch_eeg(raw) for raw in preprocessed_healthy_eeg]
schizophrenic_epochs = [epoch_eeg(raw)
                        for raw in preprocessed_schizophrenic_eeg]
print(len(healthy_epochs), len(schizophrenic_epochs))
```

```python
import numpy as np 
from Modulator import modulation 
```

```python
channels = healthy_epochs[0].ch_names
modulated_healthy_epochs = modulation(healthy_epochs, channels)
modulated_schizophrenic_epochs = modulation(schizophrenic_epochs, channels, receptor_activity= {
    '5-HT1A': {'frontal': 0.5, 'mPFC': 0.7},
    '5-HT2A': {'frontal': 0.8, 'mPFC': 0.6}
})
```

```python
def recreate_epochs(modulated_data, original_epochs):
    return mne.EpochsArray(modulated_data, original_epochs.info, original_epochs.events)


recreated_healthy_epochs = [recreate_epochs(
    mod_data, orig_ep) for mod_data, orig_ep in zip(modulated_healthy_epochs, healthy_epochs)]
recreated_schizophrenic_epochs = [recreate_epochs(
    mod_data, orig_ep) for mod_data, orig_ep in zip(modulated_schizophrenic_epochs, schizophrenic_epochs)]
```

```python
channel_names_starting_with_F = [
    ch_name for ch_name in healthy_epochs[0].ch_names if ch_name.startswith('F')]
print(channel_names_starting_with_F)
```

```python
def compare_diff(original, modulated):
    return original.get_data() - modulated


channel_names = channel_names_starting_with_F
pre_modulated = modulated_healthy_epochs[0]
pre_difference = compare_diff(healthy_epochs[0], recreated_healthy_epochs[0])
```

```python
channel_to_region = {
    'Fp1': 'Left Prefrontal', 'Fp2': 'Right Prefrontal',
    'F7': 'Left Frontal', 'F8': 'Right Frontal', 'F3': 'Left Frontal', 'F4': 'Right Frontal',
    'Fz': 'Frontal Midline', 'FC5': 'Left Fronto-Central', 'FC6': 'Right Fronto-Central',
    'FC1': 'Left Fronto-Central', 'FC2': 'Right Fronto-Central',
    'T7': 'Left Temporal', 'T8': 'Right Temporal', 'C3': 'Left Central', 'C4': 'Right Central',
    'Cz': 'Central Midline', 'CP5': 'Left Centro-Parietal', 'CP6': 'Right Centro-Parietal',
    'CP1': 'Left Centro-Parietal', 'CP2': 'Right Centro-Parietal',
    'P7': 'Left Parietal', 'P8': 'Right Parietal', 'P3': 'Left Parietal', 'P4': 'Right Parietal',
    'Pz': 'Parietal Midline', 'O1': 'Left Occipital', 'O2': 'Right Occipital'
}
region_names = [channel_to_region.get(ch, ch) for ch in channel_names]
region_names
```

```python
import matplotlib.pyplot as plt 


def plot_eeg_signals(original, modulated, difference, channel_names, pre_modulated=None, pre_difference=None,title=None):
    time_points = np.arange(original.shape[2])
    plt.figure(figsize=(22, 18))

    all_handles = []
    all_labels = []
    for i, channel in enumerate(region_names):
        plt.subplot(len(channel_names), 1, i + 1)
        h1 = plt.plot(
            time_points, original[0, i], label='Original', color='blue')
        h2 = plt.plot(
            time_points, modulated[0, i], label='Modulated (Normalized)', color='green', linestyle='--')
        h3 = plt.plot(
            time_points, difference[0, i], label='Difference (Normalized)', color='red', linestyle=':')

        if pre_modulated is not None and pre_difference is not None:
            h4 = plt.plot(
                time_points, pre_modulated[0, i], label='Pre-Modulated', color='orange', linestyle='-.')
            h5 = plt.plot(
                time_points, pre_difference[0, i], label='Pre-Difference', color='purple', linestyle='-.')

        plt.title(f'{channel}')

        if i == 0:
            all_handles.extend([h1[0], h2[0], h3[0]])
            all_labels.extend(
                ['Original', 'Modulated (Normalized)', 'Difference (Normalized)'])

            if pre_modulated is not None and pre_difference is not None:
                all_handles.extend([h4[0], h5[0]])
                all_labels.extend(['Pre-Modulated', 'Pre-Difference'])
    if title:
        plt.suptitle(title, fontsize=20)
    plt.legend(all_handles, all_labels, loc='best')
    plt.show()
```

```python

plot_eeg_signals(
    healthy_epochs[0].get_data(),
    modulated_healthy_epochs[0],
    compare_diff(healthy_epochs[0], recreated_healthy_epochs[0]),
    channel_names,
    title = 'Healthy')
```

```python
plot_eeg_signals(
    schizophrenic_epochs[0].get_data(),
    modulated_schizophrenic_epochs[0],
    compare_diff(schizophrenic_epochs[0], recreated_schizophrenic_epochs[0]),
    channel_names,
    title='Schizophrenic')
```

```python
n_reservoir = 200  
sparsity = 0.2  
spectral_radius = 1.2  

esn = ESN(n_inputs=len(channels), 
          n_outputs=19,                 
          n_reservoir=n_reservoir, 
          sparsity=sparsity, 
          spectral_radius=spectral_radius)
```

```python
def reshape_for_esn(data):
    total_samples = data.shape[0] * data.shape[2]
    reshaped_data = data.transpose(0, 2, 1).reshape(
        total_samples, data.shape[1])
    return reshaped_data  
```

```python
X_train_ = reshape_for_esn(modulated_healthy_epochs[0])
```

```python
Y_train_ = np.zeros_like(X_train_)
Y_train_[:-1] = X_train_[1:]  
Y_train_[-1] = X_train_[-1]
print(Y_train_.shape)
```

```python
esn.fit(X_train_,Y_train_)
```

```python
X_test_ = reshape_for_esn(modulated_healthy_epochs[0])
```

```python
Y_pred_ = esn.predict(X_test_)
```

```python
plt.plot(X_test_[:, 0], label="Actual")
plt.plot(Y_pred_[:, 0], label="Predicted", linestyle="--")
plt.title("Serotonin Modulation Prediction for Schizophrenic Patient")
plt.legend()
plt.show()
```

```python
def plot_eeg_signals_fixed_slice(original, modulated, difference, channel_names, predicted=None, title=None):
    time_points = np.arange(original.shape[2])  # Time points for original data
    plt.figure(figsize=(22, 18))

    all_handles = []
    all_labels = []

    for i, channel in enumerate(channel_names):
        plt.subplot(len(channel_names), 1, i + 1)

        h1 = plt.plot(
            time_points, original[0, i], label='Original', color='blue')
        h2 = plt.plot(
            time_points, modulated[0, i], label='Modulated (Normalized)', color='green', linestyle='--')
        h3 = plt.plot(
            time_points, difference[0, i], label='Difference (Normalized)', color='red', linestyle=':')
        if predicted is not None:
            predicted_modulated_slice = predicted[0][500:600, i]
            predicted_slice = predicted[1][500:600,i]
            predicted_time_points = np.arange(500, 600)
            h4 = plt.plot(predicted_time_points, predicted_slice,
                          label='Predicted', color='magenta', linestyle='--')
            h5 = plt.plot(predicted_time_points, predicted_modulated_slice ,
                          label = 'Predicted modulated',color='black',linestyle='--')

        plt.title(f'{channel}')

        if i == 0:
            all_handles.extend([h1[0], h2[0], h3[0]])
            all_labels.extend(
                ['Original', 'Modulated (Normalized)', 'Difference (Normalized)'])
            if predicted is not None:
                all_handles.extend([h4[0]])
                all_labels.extend(['Predicted'])
                all_handles.extend([h5[0]])
                all_labels.extend(['Predicted modulated'])

    if title:
        plt.suptitle(title, fontsize=20)

    plt.legend(all_handles, all_labels, loc='best')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
